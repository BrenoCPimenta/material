```python
# pop 
pop = np.round(np.random.beta(4, 2.5, size=20000) * 100)
```

```python
plt.title('Notas por Disciplinas na UFMG')
plt.hist(pop, edgecolor='k', bins=20);
```

```python
sample_size = 50
```

```python
sample_means = []
for _ in range(1000):
    sample = np.random.choice(pop, size=sample_size)
    sample_means.append(sample.mean())
```

```python
plt.title('Sample Distribution')
plt.hist(sample_means, edgecolor='k', bins=20);
```

```python
one_sample = np.random.choice(pop, size=sample_size)
sample_mean = one_sample.mean()
sample_sd = one_sample.std(ddof=1)
```

```python
estimator_sd = sample_sd / np.sqrt(sample_size)
```

```python
x = np.linspace(55, 68, 100)
x
```

```python
norm = ss.distributions.norm(one_sample.mean(), estimator_sd)
y = norm.pdf(x)
```

```python
plt.plot(x, y)
```
